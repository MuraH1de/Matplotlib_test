```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 4, 6, 12]

fig, ax = plt.subplots()

fig.suptitle('OBJECT')
#suptitleメソッドでタイトルをつけられる

ax.set_title('sub_plot')
#set_titleメソッドでもタイトルをつけられる

ax.plot(x, y)
plt.show
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 4, 6, 12]

fig, ax = plt.subplots()

fig.suptitle('OBJECT')
#suptitleメソッドでタイトルをつけられる

ax.set_title('sub_plot')
#set_titleメソッドでもタイトルをつけられる

ax.plot(x, y, label='legend label')
#label引数は凡例名を指定する
ax.legend(loc='best')
#legendメソッドで凡例を表示させる
#loc='best'で凡例の位置が一番グラフと重ならない位置に設定できる
plt.show
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 4, 6, 12]

fig, ax = plt.subplots()

fig.suptitle('OBJECT')
ax.set_title('sub_plot')

ax.plot(x, y, label='legend label')
ax.legend()
fig.savefig('safe_fiture.png')
#savefigメソッドでグラフを保存できる
```
